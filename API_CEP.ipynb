## Transformando os dados CEPS através de expressão regular

```python
import re

# A string que contém o CEP
texto = "20090-002 - Rio de Janeiro, RJ"

# Expressão regular para encontrar o CEP
padrao = r'\b\d{5}-\d{3}\b'

# Procurar o CEP na string
resultado = re.search(padrao, texto)

# Verificar se o CEP foi encontrado
if resultado:
    cep_encontrado = resultado.group()
    print("CEP encontrado:", cep_encontrado)
else:
    print("CEP não encontrado na string")
```

```python
import re
ceps=[]


for indice,dado in df.iterrows():
  dado=str(dado)

  # Expressão regular para encontrar um CEP brasileiro
  padrao_cep = r'\b\d{5}-\d{3}\b'

  # Procurar o CEP na string
  resultado = re.search(padrao_cep, dado)
  # Verificar se o CEP foi encontrado
  if resultado:
      cep_encontrado = resultado.group()
      ceps.append(cep_encontrado)
  else:
      print("CEP não encontrado na string")


```

```python
ceps
```

```python
import requests

cep_tratado=[]

dic_cep=[]

for cep in ceps:
    new_cep=cep.replace("-", "").replace(".", "").replace(" ", "")
    cep_tratado.append(new_cep)

for cep in cep_tratado:
    if len(cep) == 8:
        link = f'https://viacep.com.br/ws/{cep}/json/'

        requisicao = requests.get(link)

        print(requisicao)

        dic_requisicao = requisicao.json()

        #uf = dic_requisicao['uf']
        #cidade = dic_requisicao['localidade']
        #bairro = dic_requisicao['bairro']

        dic_cep.append(dic_requisicao)
        print(dic_requisicao)
    else:
        print("CEP Inválido")
```

```python
dados_cep=pd.DataFrame([dic_cep])
```

```python
import matplotlib.pyplot as plt

# Contar a frequência de cada valor na coluna 'uf'
contagem_uf = dados_cep['uf'].value_counts()

# Criar um gráfico de donut
fig, ax = plt.subplots()
ax.pie(contagem_uf, labels=contagem_uf.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))

# Desenhar um círculo no centro para criar o donut
centro_circulo = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centro_circulo)

# Adicionar um título
ax.set_title("Distribuição por UF")

# Exibir o gráfico
plt.axis('equal')  # Garante que o gráfico seja circular
plt.show()



```
