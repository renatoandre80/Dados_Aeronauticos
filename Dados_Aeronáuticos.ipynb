```python
import pandas as pd
```

```python
df=pd.read_csv('/content/fator_contribuinte.csv',sep=';')
```

```python
df
```

```python
df.info()
```

```python
count_fator_operacional=0

for fator_operacional in df['fator_area']:
  if fator_operacional=='FATOR OPERACIONAL':
    count_fator_operacional=count_fator_operacional+1
print(f'A quantidade de acidentes por fator operacional é {count_fator_operacional}')

```

```python
count_fator_humano=0
for fator_humano in df['fator_area']:
  if fator_humano=='FATOR HUMANO':
    count_fator_humano=count_fator_humano+1
print(f'A quantidade de acidentes por fator humano é {count_fator_humano}')


```

```python
tipos_de_fatores=df.groupby(by='fator_area').size()
tipos_de_fatores
```

```python
#PLotando gráficos:

tipos_de_fatores=df.groupby(by='fator_area').size().sort_values()[-5:].plot(kind='bar')
```

```python
count_fadiga=0
for fadiga in df['fator_nome']:
  if fadiga=='FADIGA':
    count_fadiga+=1
print(f'A quantidade de acidentes pelo fator de fadiga foi de {count_fadiga}')
```
